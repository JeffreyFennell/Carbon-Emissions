```python
import pandas as pd
import matplotlib.pyplot as plt
```

United States Carbon Emissions

```python
df = pd.read_csv('/Users/jeffreyfennell/Desktop/Coding/Git/DA-Unnamed-Project/Alternate_Data_Project/assets/co2_emissions_kt_by_country.csv')
df['country_name'].info()
```

```python
df.rename(columns= {'value': 'carbon_emissions'}, inplace= True)
```

```python
# Create df focused solely on USA data
ndf= df.drop(columns= 'country_code')
usa_df = ndf[ndf['country_name'].str.contains('United States')]
usa_df.head()
```

```python
usa_df.plot.line(x ='year', rot= 45, y= 'carbon_emissions', title = 'United States Carbon Emissions', figsize= (10, 5), grid= True)
```

```python
fig, ax = plt.subplots()
ax.plot(usa_df['year'], usa_df['carbon_emissions'])
ax.set_xlabel('Year')
ax.set_ylabel('Emissions  (in kilotons)')
ax.set_title("United States Carbon Emissions")
ax.ticklabel_format(style='plain')
plt.show()

```

```python
# Create df comparing major USA and Russia
global_powers = ['United States','Russia']
gp_df = ndf[ndf['country_name'].str.contains('Russia')]
gp_df
```

```python
fig, ax= plt.subplots()
ax.plot(gp_df['year'], gp_df['carbon_emissions'])
ax.set_xlabel('Year')
ax.set_ylabel('Emissions (in kilotons)')
ax.set_title("Emissions: China vs United States")
ax.ticklabel_format(style='plain')
plt.show()
```
